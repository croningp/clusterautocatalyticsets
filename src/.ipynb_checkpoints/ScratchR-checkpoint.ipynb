```r
library(ggplot2)

reduced_mass<-  function(m1,m2){ return (m1*m2/(m1+m2))}
```

```r
m2 = seq(1,100)
m1 = seq(1,100)
data_df = data.frame(m1 = c(), m2 = c(), rm = c())
for (one in m1){
    for (two in m2){
        data_df = rbind(data_df, data.frame(m1 = one, m2 = two, rm = reduced_mass(one,two)))
    }
}
```

```r
p <- ggplot(data_df) + geom_tile(aes(x= m1, y= m2, fill = rm)) + theme_minimal()
```

```r
p
```

```r

false_pos = data.frame(delta = c(), k = c(), fp = c())
detect_change <- function(time_series, k){
    time_series = time_series[!is.na(time_series)]
    dt = diff(time_series)
    n = length(dt)
    test_point = dt[n]
    k_series = dt[(n-k-1):(n-1)]
    k_mean = mean(k_series)
    k_std = sqrt(var(k_series))
    change_time = FALSE

    if (test_point >= k_mean + 2*k_std){
        change_time = TRUE
        }
    else if (test_point <= k_mean - 2*k_std) {
        change_time = TRUE
    }
    
    return(change_time)
}
```

```r
n_reps = 100
for (k in 5:20){
    for (delta in 2:200){
        phase1 = 600
        phase2 = 500

        half_t= 100
        max_t = 2*half_t
        
        time1 = rep(phase1, half_t)
        time2 = rep(phase2, half_t)

        time1 = time1 + sample( (phase1-delta):(phase1+delta),half_t, replace = T)
        time2 = time2 + sample( (phase2-delta):(phase2+delta),half_t, replace = T)
        time = c(time1, time2)

        fp = 0
        for (n in (k+2):max_t){
            if (detect_change(time[1:n], k)){
                fp = fp +1
            }    
        }
        fp = fp/(max_t- k+2)
        false_pos = rbind(false_pos, data.frame(delta = delta, k = k, fp = fp))
    }
}
```

```r
library(ggplot2)
```

```r
p <- ggplot(false_pos) + geom_tile(aes(x = k, y = delta, fill = fp))
```
